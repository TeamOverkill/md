```python
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np, pandas as pd
from math import sqrt, pi, exp
```

```python
x = np.loadtxt('rdf.txt', usecols=0, unpack=True)
y = np.loadtxt('rdf.txt', usecols=1, unpack=True)
```

```python
plt.plot(x,y)
plt.xlabel('$r$ / nm')
plt.ylabel('$g(r)$')
plt.savefig('rdf.pdf')
```

```python
conc = y*0.14
print(conc)
plt.plot(x, conc)
dx = x[1]-x[0]
vol = np.sum(4*np.pi*x**2*dx)
print("dx: ", dx)
print("Volume: ", vol)
N_sum=4*np.pi*np.cumsum(conc*x**2)*dx
plt.plot(x,N_sum)
plt.xlabel('$r$ / nm')
plt.ylabel('$N_{Cl^-}$')
```
